```python
#define a linear model
def model(X,Y,learning_rate,iteration):
    m=Y.size
#     define theta as a vector of zeros, matrix of size n,1
    theta = np.zeros((X.shape[1],1))
    cost_list=[]
    for i in range (iteration):
        y_pred=np.dot(X, theta)
#         formula cost function
        cost = (1/(2*m))*np.sum(np.square(y_pred - Y))
#     what is dtheta
        d_theta = (1/m)*np.dot(X.T, y_pred - Y)
        theta = theta - learning_rate*d_theta
        cost_list.append(cost)
        if(i%(iteration/10)==0):
            print("Cost is:", cost)
        
    return theta,cost_list
print("hehe")
    
```

```python
iteration = 900
learning_rate = 0.000000000015
theta,cost_list = model(X, Y, learning_rate = learning_rate,iteration = iteration)
```

```python
# plot graph for cost and iteration
rng = np.arange(0,iteration)
plt.plot(rng, cost_list)
plt.xlabel('Cost')
plt.ylabel('No. of iterations')
plt.show()
```
